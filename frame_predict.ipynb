```python
source_dir =  "frames/"
prediction_folder="prediction/"
def images_class(list):
    car=0
    heavy=0
    two_wheel=0
    position = (10,50)
    for name in list:
        path = source_dir+name
        predict=recognize_image(path)
        for object in prediction[0]["predictions"]["predictions"]:
            if (object["class"]=="car"):
                car =car+1
            elif (object["class"]=="heavy"):
                heavy =heavy+1
            elif (object["class"]=="two-wheel"):
                two_wheel= two_wheel+1
            text="Car: "+str(car)+"| Heavy: "+str(heavy)+"| Two_wheel: "+str(two_wheel)
        encode=predict["output_image"]
        head,tail=os.path.splitext(name)
        target = prediction_folder+"."+head + "_prediction" + tail
        imgdata = base64.b64decode(encode)
 # I assume you have a way of picking unique filenames
        with open(target, 'wb') as f:
            f.write(imgdata)
        image = cv2.imread(target)
        cv2.putText(image,text=text,org=position,fontFace= cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(209, 80, 0, 255),
            thickness=1, lineType=cv2.LINE_AA)
        cv2.imwrite(target, image)
        car=0
        heavy=0
        two_wheel=0
```
